```python
from pycaret.datasets import get_data
```

```python
data1 = get_data(dataset="traffic")
```

```python
from pycaret.regression import *
```

```python
dataset = setup(data=data1, target="traffic_volume",session_id=438,imputation_type='simple')
```

```python
best = compare_models(include=['lr', 'lasso', 'ridge','en','lar'	,'llar'	,'omp'
                               ,'br'	,'ard'	,'par'])
```

```python
tuned_best = tune_model(estimator = best)
```

```python
# models()
```

```python
final_model = finalize_model(estimator=tuned_best)
```

```python
save_model(model = final_model,model_name="./rf_base_traffic")
```

```python
saved_model = load_model("./rf_base_traffic")
```
